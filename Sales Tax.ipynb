```python
def func(l):
    notax = ['books','book','food','medical','chocolate','headache','chocolates']
    temp = []
    res = []
    flag = False
    ob=float(0)
    per1=float(0)
    per2=float(0)
    tax=float(0)
    total=float(0)
    lt=float(0)
    for i in l:
        temp = i.split()
        for j in temp:
            if j in notax:
               flag=True
               lt=float(temp[len(temp)-1])
               lt = round(lt,2)
               break
            else:
                lt=float(temp[len(temp)-1])
                lt = round(lt,2)
        ob = ob + float(lt)
        if flag:
            if 'imported' in temp:
                per1 = (float(temp[0])*float(lt)*5)/100
                tax = tax+per1
                total = (float(temp[0])*float(lt))+per1
                total = round(total,2)
                res.append(total)
            else:
                res.append(float(lt))
        else:
            if 'imported' in temp:
                per1=(float(temp[0])*float(lt)*5)/100
                per2=(float(temp[0])*float(lt)*10)/100
                tax = tax+per1+per2
                total=(float(temp[0])*float(lt))+per1+per2
                total = round(total,2)
                res.append(total)
            else:
                per2=(float(temp[0])*float(lt)*10)/100
                tax = tax+per2
                total=(float(temp[0])*float(lt))+per2
                total = round(total,2)
                res.append(total)
        flag = False
    ind=0
    res1=[]
    nt = float(0)
    for i in l:
        temp = i.split()
        temp[len(temp)-1] = str(res[ind])
        ind=ind+1
        x = ' '.join(temp)
        res1.append(x)
    for i in res:
        nt = nt+i
    print()
    print("Without tax: ")
    for i in l:
        print(i)
    print()
    print("With tax: ")
    for i in res1:
        print(i)
    print()
    print("Total Price without tax: ",round(ob,2))
    print("Sales Tax: ",round(tax,2))
    print("Total Price: ",nt)

l=[]
n = int(input("Enter the number: "))
print()
for i in range(n):
    s = input("Enter the string: ")
    l.append(s)
func(l)
```
